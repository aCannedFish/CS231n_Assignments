**Inline Question 2**

Although gradcheck is reliable softmax loss, it is possible that for SVM loss, once in a while, a dimension in the gradcheck will not match exactly. What could such a discrepancy be caused by? Is it a reason for concern? What is a simple example in one dimension where a svm loss gradient check could fail? How would change the margin affect of the frequency of this happening?

Note that SVM loss for a sample $(x_i, y_i)$ is defined as: $$L_i = \sum_{j\ne y_i}\max(0, s_j - s_{y_i} + \Delta)$$ where $j$ iterates over all classes except the correct class $y_i$ and $s_j$ denotes the classifier score for $j^{th}$ class. $\Delta$ is a scalar margin. For more information, refer to 'Multiclass Support Vector Machine loss' on [this](https://cs231n.github.io/linear-classify/) page.

*Hint: the SVM loss function is not strictly speaking differentiable.*


$\color{blue}{\textit Your Answer:}$ 


可能的原因：
- SVM（hinge）损失在边界点不可导。具体地，单样本对某一不正确类别的项是 max(0, s_j - s_y + Δ)。当其内部表达式 z = s_j - s_y + Δ 恰好等于 0 时，函数在该点处不可导（左导为 0，右导为 1）。数值梯度（finite difference，通常是中心差分）会逼近左右导数的平均或受 ε 方向影响，从而可能与你在代码里选取的解析（子）导数不一致。
- 数值精度与差分步长 ε 的选择也会导致小差异：ε 太大或太小都会使数值近似偏离解析值或受舍入误差影响。
- 极少数情况下，标签或实现错误也会造成一致性问题，但如果只有个别维度在特定点不匹配，最常见的原因是上面两个（不可导或数值误差）。

这是否值得担心：
- 如果只有少数维度在边界点发生不匹配，通常不是问题：这是 hinge 的数学特性（非光滑点）导致的预期行为，不代表实现错误。
- 如果大量或多数维度不匹配，则要怀疑实现错误（比如索引、广播、正则项或损失/梯度公式写错），这时就需要调试。

一维简单例子（手算能看出不匹配）：
- 设二分类，正确类为 y=0，两个分数 s_y = 0, s_j = -1，设 Δ = 1。
- 对该不正确类：z = s_j - s_y + Δ = -1 - 0 + 1 = 0，处于 hinge 的拐点。
- 此时单项损失为 max(0, z) = 0。对 z 的左导数是 0（如果把 s_j 稍减，z<0, 导数为0），右导数是 1（若把 s_j 稍增，z>0, 导数为1）。中心差分近似可能给出 ~0.5（或其他取决于实现与 ε），而实现中通常选用的子导数可能是 0 或 1，所以 gradcheck 会报告不匹配。

关于 margin Δ 如何影响发生频率：
- Δ 会移动“等于零”的边界位置（z = s_j - s_y + Δ = 0 的超平面）；改变 Δ 会改变哪些样本处于该边界，但并不会消除非可导性的本质。
- 一般来说，随机连续分数精确落在等于零的位置概率几乎为 0，因此这种数值不匹配比较罕见。增大或减小 Δ 会改变处于“恰好等于零”条件的样本集合，但不会改变“只有在边界点才可能出现不匹配”的结论。
- 实际上，增大 Δ 会使更多样本满足 z>0（活跃 margin），但恰等于 0 的情形仍然是边界情况，频率主要受分数分布和数值精度影响。

实用建议：
- 如果遇到单个维度不匹配且怀疑是边界导致，可把 gradcheck 的差分步长 ε 改小/改大试试，或把那一维排除再观察整体匹配情况。
- 对于调试，打印出导致不匹配的样本索引和值（scores, z）通常能快速定位是“恰好在边界”还是实现错误。

**Inline Question 4** - *True or False*

Suppose the overall training loss is defined as the sum of the per-datapoint loss over all training examples. It is possible to add a new datapoint to a training set that would change the softmax loss, but leave the SVM loss unchanged.

$\color{blue}{\textit Your Answer:}$ True

$\color{blue}{\textit Your Explanation:}$ 可以。SVM hinge 损失对单个样本是 max(0, s_j - s_y + Δ) 的和：如果新样本已经满足对所有 j≠y 都有 s_y ≥ s_j + Δ，则该样本的 SVM 损失为 0，把它加到训练集里不会改变“损失总和”（增加 0）。但 Softmax 交叉熵是 -log p_y，只要 p_y 不是严格等于 1（通常不会），它就是一个正数，因此加入该样本会让 softmax 总损失发生变化。
