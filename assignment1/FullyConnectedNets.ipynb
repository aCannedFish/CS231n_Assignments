Did you notice anything about the comparative difficulty of training the three-layer network vs. training the five-layer network? In particular, based on your experience, which network seemed more sensitive to the initialization scale? Why do you think that is the case?

## Answer:
一般来说五层网络比三层网络更难训练、对初始化尺度（weight scale）也更敏感。原因是前向与反向传播都要经过更多层：
- 如果初始化过小，每层输出的方差会逐层缩小，信号与梯度在传播过程中容易衰减（vanishing），导致学习非常慢，loss/acc 变化不明显。
- 如果初始化过大，激活与梯度的方差会被逐层放大（exploding），训练不稳定；对 ReLU 还可能出现大量神经元长期输出 0（“dead ReLU”），同样会影响收敛。
- 更深的网络还会带来更差的条件数/优化更困难，因此对一个合适的初始化方差范围更“挑剔”。
三层网络层数少，上述累积效应较弱，所以在相同的超参搜索下通常更容易找到能收敛甚至过拟合的小范围初始化尺度。

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
在 AdaGrad 里，`cache` 会把每个参数历史上的梯度平方不断累加：`cache_t = \sum_{\tau=1}^t dw_\tau^2`。因此随着训练进行，`cache` 单调变大，分母 `\sqrt(cache) + eps` 也越来越大，导致有效步长变成
$$\alpha_{\text{eff}}(t) = \frac{\text{learning\_rate}}{\sqrt{cache_t}+\epsilon}$$
会不断变小，最后更新幅度非常小（尤其是在早期梯度较大或噪声较多时更明显），从而学习变慢甚至“停滞”。

Adam 一般不会像 AdaGrad 那样严重，因为它用的是梯度一阶/二阶矩的**指数滑动平均**（EMA），而不是把平方梯度永远累加；二阶矩 `v_t` 主要反映“近期”的梯度尺度，并且有 bias correction，所以不会出现分母无界、步长持续衰减到几乎为 0 的同样问题。
不过 Adam 仍然会做自适应缩放：如果某些参数的近期梯度方差很大，`\sqrt{v_t}` 也会变大，更新也会相对变小，但通常不会像 AdaGrad 那样单调地越训越小。
